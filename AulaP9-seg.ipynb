## Criação dum dataset em texto (exercício 2)

Cria um ficheiro de texto com meia dúzia de entradas.

A seguir especifica a função que o carrega para o modelo em Python.

```python
# Leitura/carregamento da informação do ficheiro
import json

def chaveOrd(entrada):
    return entrada['data']

def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    dicJSON=json.load(f)
    bd=dicJSON['diario']
    #Acrescentar o campo em falta (id) e remover o dado 'nr_para'
    i=1
    for entrada in bd:
        entrada['id']='e'+str(i)
        del entrada['nr_para']
        i=i+1
    #Ordenar as entradas por ordem cronológica decrescente
    bd.sort(key=chaveOrd, reverse=True)
    return bd

Diario = []
Diario = lerDataset("../AGTP/diario2.json")
print(Diario)
```

```python
import json

def guardarBD(bd,fnome):
    f=open(fnome,'w',encoding='utf-8')
    json.dump(bd,f,ensure_ascii=False,indent=4)
    
guardarBD(Diario,'myDiary.json')

def carregarBD(fnome):
    bd=[]
    f=open(fnome,encoding='utf-8')
    bd=json.load(f)
    return bd

AulaTP=carregarBD('../AGTP/myDiary.json')
print(AulaTP)
```

## Interface (exercício 3)

Especifica uma interface para a tua aplicação onde seja possível visualizar as operações possíveis e as últimas 10 entradas no diário (ordem cronológica decrescente).

```python
# Interface
import PySimpleGUI as sg

def entradasListBox(bd):
    entradas=[]
    for entrada in bd:
        if 'descrição' in entrada:
            entradas.append(entrada['data']+': '+entrada['descrição'])
        else:
            entradas.append(entrada['data']+': ------')
    return entradas

# Painel com duas colunas: 
# Coluna 1: menu
# Coluna 2: Dados

menu_list_column = [
    [sg.Button("Limpar")],
    [sg.Button("Carregar BD")],
    [sg.Button("Gravar BD")],
    [sg.Button("Listar BD")],
    [sg.Button("Inserir registo")],
    [sg.Button("Sair")]
]

listaEntradas=entradasListBox(AulaTP)      

data_viewer_column = [
    [sg.Text("Painel de Dados")],
    [sg.Listbox(values=entradas, size=(20,10), key='_entradas_')],
]

interfacePrincipal = [
    [
        sg.Column(menu_list_column),
        sg.VSeperator(), #VSeperator - separador vertical: inserção de uma lista
        sg.Column(data_viewer_column),
    ]
]


window=sg.Window(title='O meu diário', font='Helvetica 12', default_element_size=(20,1)).Layout(interfacePrincipal)

stop=False
while not stop:
    event, values = window.read()
    if event == sg.WIN_CLOSED or event =='Sair':
        guardarBD(AulaTP, 'myDiary.json')
        stop=True
    elif event == 'Limpar':
        AulaTP.clear()
        window.find_element('_entradas_').update(values=[])
    elif event == 'Carregar':
        AulaTP=carregarBD('myDiary.json')
        listaEntradas=entradasListBox(AulaTP) 
        window.find_element('_entradas_').update(values=listaEntradas)
    elif event == 'Inserir':
        Pass
    else:
        print(event + ':\n')
        print(values)
        #print(event+ ':\n' + '\n'.join('{}\t{}'.format(k,v) for k,v in values.items())) #join - imprime o elemento em cada linha
window.close()
```

9.2. Qual a distribuição dos indivíduos por profissão?

```python
def distProfissao(bd): 
    res = {}
    
    for individuo in bd:
        profiss = individuo['profissao']
        
        if profiss in res.keys(): 
            res[profiss] += 1

        else: 
            res[profiss] = 1

    prof = sorted(res.items(), key=lambda x:x[1])
    return prof

distProfissao(BD)
```

```python
#Função auxiliar da distribuição em plot.

def top20Prof(bd):
    res = distProfissao(bd)

    top10 = res[-20:]

    return top10

print(top20Prof(BD))
```

9.3. Qual a distribuição de indivíduos por desporto (apenas o top10)?

```python
def distDesporto(bd): 
    dist = {}

    for indiv in bd: 
        for desporto in indiv['desportos']:

            if desporto in dist.keys(): 
                dist[desporto] += 1

            else: 
                dist[desporto] = 1
    
    desportos = sorted(dist.items(), key=lambda x:x[1])
    desportoss = desportos[-10:]

    return desportoss


distDesporto(BD)
```

```python
def janelaErro(mensagem):
    interface = [
        [sg.Text(mensagem,text_color='peru',background_color='beige')],
        [sg.Button('Ok',button_color='peru')]
    ]
    
    window = sg.Window(title='ERRO',background_color = 'beige').Layout(interface)
    
    stop=False
    while not stop:
        event, values = window.read()
        if event == 'Ok' or event == sg.WIN_CLOSED:
            stop=True
    
    window.close()
    exit()
```

```python
import os

def adicionarDataset(BD):

    file_list_column = [
        [
            sg.Text('Pasta',font='Arial 10'),
            sg.In('Ficheiro selecionado', size=(20,1),enable_events=True, font='Arial 10',key="-PASTA-"),
            sg.FolderBrowse(font='Arial 10',size=(5,2))
        ],
        [
            sg.Listbox(values=[], enable_events=True, font='Arial 10',size=(25,30),key="-LISTA-")
        ]]

    file_viewer_column = [ 
        [sg.Text("Pasta selecionada:", font='Arial 10')],
        [sg.Text("Nome do ficheiro:", font='Arial 10'),sg.Text(key='-NOME-',font='Arial 10')],
        [sg.Multiline( key="-CONTEUDO-", font='Arial 10',size=(25,30))],
        [sg.Button("Adicionar ao Dataset",font='Arial 10',size=(5,2)),sg.Button("Sair",font='Arial 10',size=(5,2))]
    ]

    layout = [[
        sg.Column(file_list_column),
        sg.VSeparator(),
        sg.Column(file_viewer_column)
    ]]

    window = sg.Window("Visualizador de ficheiros", margins=(100,50), font=('Arial', 24)).Layout(layout)

    folder_location = ""
    stop = False

    while not stop:
        event, values = window.read()

        if event in [sg.WIN_CLOSED, "EXit"]:
            stop = True
        elif event == "-PASTA-":
            folder_location = values["-PASTA-"]
            try:
                files = os.listdir(folder_location)
            except:
                files = []

            filenames = [
                file for file in files
                if os.path.isfile(os.path.join(folder_location, file))
                and file.lower().endswith((".txt", ".csv", ".json", ".py"))
            ]
            window["-LISTA-"].update(filenames)

        elif event == "-LISTA-" and len(values["-LISTA-"]) > 0:
            file_selection = values["-LISTA-"][0]
            file = open(os.path.join(folder_location, file_selection))
            contents = file.read()
            window["-NOME-"].update(os.path.join(folder_location, file_selection))
            window["-CONTEUDO-"].update(contents)

        elif event == "Adicionar ao Dataset" and len(values["-LISTA-"]) > 0:
            file_selection = values["-LISTA-"][0]
            acrescentarBD(file_selection)

    window.close()
    exit()
```

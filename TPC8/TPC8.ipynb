### Resumo

Durante a oitava aula prática, foi proposta aos alunos a criação de uma app que permita ler e iterar sobre um dataset de EMDs. Para auxiliar, foi criado um módulo composto pelas funções criadas com a ajuda do supervisor durante a aula.

A aplicação é uma interface composta por duas colunas: uma composta por botões que indicam as possíveis funcionalidades da aplicação e outra por um painel de dados. O utilizador, portanto, pode carregar e listar a BD, e, também, conferir as diferentes distribuições dos EMDs registados no ficheiro (ano, aptidão, modalidade, clube, idade, federação, género).

No painel de dados, são apresentadas as opções 'Inserir novo EMD' e 'Remover EMD', o que permite o utilizador alterar os dados da Base de Dados. Para tornar o processo visualmente melhor e mais prático, ambos os botões abrem uma nova janela que permite realizar as determinadas funções.



```python
import modulo8 as emd
import PySimpleGUI as sg
```

```python
def janelaInserir(d):
    sg.theme('Reddit')
    interface=[
        [sg.Text('emd'+str(len(d)+1),key='-emd')],
        [sg.Input('Data: AAAA-MM-DD', key='-data')],
        [sg.Input('Primeiro nome', key='-firstname'), sg.Input('Último nome', key='-lastname')],
        [sg.Input('Idade', key='-age'), sg.Input('Género: M ou F', key='-genero')],
        [sg.Input('Morada', key='-morada')],
        [sg.Input('Modalidade', key='-modalidade'), sg.Input('Clube', key='-clube')],
        [sg.Input('Email', key='-email')],
        [sg.Input('Federado: true ou false', key='-federado'), sg.Input('Resultado: true ou false', key='-resultado')],
        [sg.Button('Inserir novo EMD')]
        ]
    window=sg.Window('Inserir').Layout(interface)
    stop=False
    while not stop:
        event, values = window.read() 
        if event == sg.WIN_CLOSED: 
            stop = True
        elif event == 'Inserir novo EMD':
            novoEMD=[]
            data=values['-data']
            firstname=values['-firstname']
            lastname=values['-lastname']
            age=values['-age']
            genero=values['-genero']
            morada=values['-morada']
            modalidade=values['-modalidade']
            clube=values['-clube']
            email=values['-email']
            federado=values['-federado']
            resultado=values['-resultado']
            novoEMD.append('emd'+str(len(d)+1))
            novoEMD.append(data)
            novoEMD.append(firstname)
            novoEMD.append(lastname)
            novoEMD.append(age)
            novoEMD.append(genero)
            novoEMD.append(morada)
            novoEMD.append(modalidade)
            novoEMD.append(clube)
            novoEMD.append(email)
            novoEMD.append(federado)
            novoEMD.append(resultado) 
            window.close()
    return novoEMD
```

```python
def janelaRemover(d):
    sg.theme('Reddit')
    interface=[
        [sg.Text('Qual emd desejas remover?'), sg.Input('emdXXX',key='-remover')],
        [sg.Button('Remover')]
    ]
    
    window=sg.Window('Remover').Layout(interface)
    stop=False
    while not stop:
        event, values = window.read() 
        if event == sg.WIN_CLOSED: 
            stop = True
        elif event == 'Remover':
            emd=values['-remover']
            for elem in d:
                if elem[0]==emd:
                    d.remove(elem)
        window.close()
    return d
    
```

```python
BD = []
BD = emd.lerDataset("../AGTP/emd.csv")

sg.theme('Reddit')
menu_list_column=[
    [sg.Button('Carregar BD')],
    [sg.Button('Listar BD')],
    [sg.Button('Dist. por Ano')],
    [sg.Button('Dist. por Aptidão')],
    [sg.Button('Dist. por Clube')],
    [sg.Button('Dist. por Federação')],
    [sg.Button('Dist. por Género')],
    [sg.Button('Dist. por Modalidade')],
    [sg.Button('Sair')]
]

data_viewer=[
    [sg.Text('Painel de Dados', font='Haveltica 14')],
    [sg.Button('Inserir novo EMD'), sg.Button('Remover EMD')],
    [sg.Output(size=(100,30),key='-output-')]
]

interface=[
    [sg.Column(menu_list_column),
    sg.VSeparator(),
    sg.Column(data_viewer)]
    
]

window=sg.Window('EMD').Layout(interface)

stop=False
while not stop:
    event,values=window.read()
    if event == 'Sair' or event == sg.WIN_CLOSED:
        stop = True
    elif event == 'Inserir novo EMD':
        novoEMD=janelaInserir(BD)
        BD.append(novoEMD)
    elif event == 'Remover EMD':
        janelaRemover(BD)
    elif event == 'Carregar BD':
        window.find_element('-output-').update(BD)
    elif event == 'Listar BD':
        window.find_element('-output-').update('')
        tabela=emd.listarDataset(BD)
        window.find_element('-output-').update(tabela)
    elif event == 'Dist. por Ano':
        BD=emd.bdOrd(BD)
        window.find_element('-output-').update('')
        ano=emd.distrib(BD,'ano')
        window.find_element('-output-').update(ano)
    elif event == 'Dist. por Aptidão':
        BD=emd.bdOrd(BD)
        window.find_element('-output-').update('')
        apto=emd.distrib(BD,'apto')
        window.find_element('-output-').update(apto)
    elif event == 'Dist. por Clube':
        BD=emd.bdOrd(BD)
        window.find_element('-output-').update('')
        clube=emd.distrib(BD,'clube')
        window.find_element('-output-').update(clube)
    elif event == 'Dist. por Federação':
        BD=emd.bdOrd(BD)
        window.find_element('-output-').update('')
        federado=emd.distrib(BD,'federado')
        window.find_element('-output-').update(federado)
    elif event == 'Dist. por Género':
        BD=emd.bdOrd(BD)
        window.find_element('-output-').update('')
        genero=emd.distrib(BD,'genero')
        window.find_element('-output-').update(genero)
    elif event == 'Dist. por Modalidade':
        BD=emd.bdOrd(BD)
        window.find_element('-output-').update('')
        modalidade=emd.distrib(BD,'modalidade')
        window.find_element('-output-').update(modalidade)
    
    
window.close()
```
